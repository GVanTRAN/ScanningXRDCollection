```python
from pathlib import Path
import re

folder = Path(r"grouped_data/dq_pap217/dq_pap217_roi146597_164520")

files = sorted([f for f in folder.iterdir() if f.is_file()])

print(f"num_runs {len(files)}")
print('out_file = Concat(String(INP_File), ".INP");')
print("#list File_Name Row Col Index {")

for f in files:
    match = re.search(r"_r(\d+)_c(\d+)_idx(\d+)", f.name)

    if match:
        row = int(match.group(1))
        col = int(match.group(2))
        idx = int(match.group(3))

        print(f"    xyfiles4\{f.name:<40} {row:<5} {col:<5} {idx:<5}")

print("}")
```
